```python
import pandas as pd

# Excel 파일 불러오기
picturethis_df = pd.read_excel('data/picturethis_15677.xlsx')

# CSV 파일 불러오기
usda_df = pd.read_csv('data/USDA_2269.csv')
picturethis_df

```

```python
# 선택하고자 하는 열 목록
selected_columns = [
    'name','latin_name','꽃 지름', '꽃 색깔', '잎 색깔', 'sunlight', '식물 높이', '수명', 'water', '이상적인 온도'
]

# 선택된 열만 포함하는 새로운 DataFrame 생성
selected_df = picturethis_df[selected_columns]
selected_df = selected_df.dropna(how='all')
# 각 행의 NaN 값의 수를 계산
nan_counts_per_row = selected_df.isna().sum(axis=1)

# NaN 값이 4개 미만인 행만 선택
rows_with_less_than_four_nans = nan_counts_per_row < 4

# 조건에 맞는 행들로 구성된 새로운 DataFrame 생성
selected_df = selected_df[rows_with_less_than_four_nans]
# 인덱스를 재할당하고, 이전 인덱스를 제거
selected_df = selected_df.reset_index(drop=True)

selected_df
```

```python
# 'ScientificName' 열에서 띄어쓰기로 단어를 나누고, 처음 두 단어를 선택하여 새로운 리스트를 생성
first_two_words = usda_df['ScientificName'].str.split().str[:2]

# 선택된 두 단어를 공백으로 다시 연결
# 'ScientificName' 열에서 처리 시작
usda_df['FirstTwoWords'] = usda_df['ScientificName'].apply(
    lambda name: ' '.join([word.lstrip('×') for word in name.split()][:])
)

usda_df['FirstTwoWords']
```

```python
# 'latin_name' 열과 'ScientificName'열의 값을 전부 소문자로 변환
selected_df['latin_name_lower'] = selected_df['latin_name'].str.lower()
usda_df['CommonName_lower'] = usda_df['CommonName'].str.lower()

# 변환된 값을 기준으로 겹치는 값 확인
overlap_bool = selected_df['latin_name_lower'].isin(usda_df['CommonName_lower'])

# 겹치는 값의 개수 계산
overlap_count = overlap_bool.sum()

# 겹치는 값의 개수 출력
print(f"겹치는 값의 개수: {overlap_count}")
selected_df['latin_name_lower']
```

```python
from fuzzywuzzy import fuzz

# 겹치는 값을 찾는 함수 정의
def find_best_match(query, choices, score_cutoff=75):
    best_score = 0
    best_match = None
    for choice in choices:
        score = fuzz.token_sort_ratio(query, choice)
        if score > best_score and score >= score_cutoff:
            best_score = score
            best_match = choice
    if best_match:
        return best_match, best_score
    else:
        return None

# usda_df의 'CommonName_lower'열의 값들
choices = usda_df['CommonName_lower'].unique()

# 각 selected_df의 'latin_name_lower' 값에 대해 가장 비슷한 값을 찾고, 그 결과를 저장
selected_df['match'] = selected_df['latin_name_lower'].apply(lambda x: find_best_match(x, choices))

# 유효한 매치가 있는 행만 필터링
valid_matches = selected_df[selected_df['match'].notna()]

# 유효한 매치의 개수 출력
print(f"유효한 매치의 개수: {len(valid_matches)}")
```

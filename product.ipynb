```python
import pandas as pd
import re
# 불용어 리스트
stopwords = ['상품', '신제품', '최고', '한정판', '제품']

# 전처리 함수 정의
def preprocess_text(text):
    # 특수 문자 및 숫자 제거
    text = re.sub(r'[^가-힣A-Za-z]', ' ', text)
    # 불용어 제거
    text = ' '.join(word for word in text.split() if word not in stopwords)
    return text
# Function to remove text within square brackets
def remove_text_within_brackets(text):
    return re.sub(r'\[.*?\]', '', text)

# Applying the function to the 'product name' column
df['상품명'] = df['상품명'].apply(remove_text_within_brackets)
# '상품명' 열에 대해 전처리 함수 적용
df['상품명'] = df['상품명'].apply(preprocess_text)

df
```

```python
# Excel 파일 불러오기
picturethis_df = pd.read_excel('data/picturethis_15677.xlsx')
```

```python
from fuzzywuzzy import process, fuzz

# 겹치는 값을 찾는 함수 정의 (fuzz.token_sort_ratio 사용)
def find_best_match(query, choices, score_cutoff=10):
    # 모든 선택지에 대한 유사도 점수를 계산
    scores = [(choice, fuzz.token_sort_ratio(query, choice)) for choice in choices]
    
    # 지정된 점수 이상의 가장 높은 점수를 가진 항목 찾기
    best_match = max(scores, key=lambda x: x[1], default=(None, 0))
    if best_match[1] >= score_cutoff:
        return best_match
    else:
        return None

# choices 정의 예시 (picturethis_df['name'].unique() 기반 가정)
choices = picturethis_df['name'].unique()

# 'latin_name_lower' 값을 기준으로 가장 유사한 값을 찾아 'match'에 저장
df['match'] = df['상품명'].apply(lambda x: find_best_match(x, choices))

# 유효한 매치가 있는 행만 필터링
valid_matches = df[df['match'].notna()]

# 유효한 매치의 개수 출력
print(f"유효한 매치의 개수: {len(valid_matches)}")
#valid_matches
```
